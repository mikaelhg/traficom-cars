```python
import pandas as pd

def tlad_5_8_conv_ensirekisterointipvm(x):
    return pd.to_datetime(x, format='%Y-%m-%d', errors='coerce')

def tlad_5_8_conv_kayttoonottopvm(x):
    return pd.to_datetime(x, format='%Y%m%d', errors='coerce')

def tlad_5_8_conv_int(x):
    if x and len(x) > 0: return int(x)
    else: return 0

TLAD_5_8_PD_READCSV = {
    'sep': ';',
    'encoding': 'iso-8859-15',
    'converters': {
        'ensirekisterointipvm': tlad_5_8_conv_ensirekisterointipvm,
        'kayttoonottopvm': tlad_5_8_conv_kayttoonottopvm,
        'Co2': tlad_5_8_conv_int,
        'matkamittarilukema': tlad_5_8_conv_int
    },
    'dtype': {
        'ajoneuvoluokka': 'category',
        'kaupallinenNimi': 'category',
        'merkkiSelvakielinen': 'category'
    },
    'usecols': [
        'ajoneuvoluokka',
        'ensirekisterointipvm',
        'kayttoonottopvm',
        'kaupallinenNimi',
        'Co2',
        'matkamittarilukema',
        'merkkiSelvakielinen',
        'mallimerkinta'
    ]
}


df = pd.read_csv('data/TieliikenneAvoinData_5_8.csv.xz',
                 engine='c',
#                 nrows=20000,
                 **TLAD_5_8_PD_READCSV)

cars = df[(df['ajoneuvoluokka'] == 'M1') & (df['Co2'] > 0) & (df['matkamittarilukema'] > 0)]

cars.sample(10)
```

```python
def malli_split(x):
    vals = x.split(' ', 2)
    ret = ['', '', x]
    if len(vals) > 1 and len(vals[0]) == 2 and vals[0][1] == 'D':
        ret = vals
    elif len(vals) == 3:
        ret = ['', vals[0], vals[1] + ' ' + vals[2]]
    elif len(vals)  == 2:
        ret = ['', vals[0], vals[1]]
    return pd.Series(ret)

if 'mallimerkinta' in cars:
    cars[['doors', 'model', 'special']] = cars.mallimerkinta.apply(malli_split)
    cars.drop(columns='mallimerkinta', inplace=True, errors='ignore')

cars.sample(100)
```
